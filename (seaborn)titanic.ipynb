```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
import numpy as np

titanic=sns.load_dataset("titanic")
print(titanic)


#graphs 
sns.displot(data=titanic , x="age",bins=10,hue="sex",col="alive",row='alone', kind="hist",color="red",edgecolor="black")
sns.displot(data=titanic , x="age", hue="alive",col="sex", kind="kde", fill=True,edgecolor="black")
# sns.regplot(data=titanic , x="survived",y="pclass")



```

```python

sns.displot(data=titanic, x="fare",fill=True ,linewidth=0.5,color="red",edgecolor="green", height=8)
#,multiple='layer'

sns.rugplot(data=titanic, x="fare", color="blue", edgecolor="black",linewidth=1)

#rug plot is axis leel function whihc is used with kde or hist to  denote exact distribution to identify outlieres , skewed , min , max etdc .it is axis level function
```

```python
sns.histplot(data=titanic,x="age",kde=True, fill=True, hue="sex")
```

```python
#categorial level 
# barplot , countplot, boxplot , strip, swarm, point,violinplot =box+kde
sns.countplot(data=titanic,x="sex",hue="survived",order=["female", "male"], palette="Set2")



```

```python
sns.catplot(data=titanic, x="pclass", y="fare",kind="bar",errorbar=None,estimator=np.max,hue="sex", palette="deep")
sns.catplot(data=titanic, x="pclass", y="fare",kind="point",errorbar=None,estimator=np.max,hue="sex")
```

```python
sns.catplot(data=titanic, x="survived", y="age",kind="bar",errorbar=None,hue="sex")
```

```python
sns.catplot(data=titanic, x="pclass", y="age",kind="box",palette="deep",hue="sex")
```

```python
sns.catplot(data=titanic,x="sex", y="fare",kind="violin",palette="coolwarm",hue="alive")
```

```python
sns.catplot(data=titanic, x="pclass", y="age", kind="strip",color="red", jitter=False)
sns.catplot(data=titanic, x="pclass", y="age", kind="swarm",color="red")
```

```python
sns.catplot(data=titanic, x="pclass",y="survived",kind="point",errorbar=None, hue="sex",palette="deep")
```

```python
sns.relplot(data=titanic,x="fare",y="age",hue="sex",style="alive",col="pclass",size="sibsp", row="embarked")

```

```python
sns.regplot(data=titanic,x="age",y="fare")
```

```python
sns.pairplot(data=titanic, hue="sex")
```

```python
g=sns.PairGrid(data=titanic, hue="sex")
g.map_offdiag(sns.barplot)
g.map_diag(sns.histplot)
g.map_upper(sns.stripplot)
g.map_lower(sns.pointplot)
```

```python
sns.lmplot(data=titanic, x="pclass",y="survived",hue="sex")
```

```python
corr = titanic.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.show()
```
