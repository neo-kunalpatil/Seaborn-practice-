```python
# matrix plot - heatmap and clustermap
import seaborn as sns
import matplotlib as plt
import plotly.express as px

gap= px.data.gapminder()

life= gap.pivot(index= "country", columns="year", values="lifeExp")
print(gap)

sns.heatmap(life)
#custommization


euro= gap[gap['continent']== "Europe"].pivot(index="country", columns="year", values="lifeExp")
sns.heatmap(euro ,annot=True, linewidth= 0.5 , cmap="summer");



```

```python
g=sns.PairGrid(data=tips, hue="sex")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.histplot)
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=tips, hue="sex", vars= ["tip","total_bill"])
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.histplot)
g.map(sns.scatterplot)
```
